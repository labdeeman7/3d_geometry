### Match each M2 to tree of M1 on other scan

```python
def match_points(tree, points):
    points_shape = points.shape
    dist, matches =  tree.query(points)
    match_information = dist, matches
    return [dist, matches]
```

### Reject “bad” pairs (pi, qi)
Rejecting points whose corresponding distance with its matched point is too large. I used the mean + multiple*std(muliple defaults to 2) of all the corresponding distances between corresponding points as a limit

```python
def reject_bad_points(dist, matches, reference_points, moving_points, boundary_ratio = 0, multiple = 2):
    '''
        function to remove the points at the boundary, and points that are larger than 2/n times the sd + mean 
    '''
    #bad points by angles. For later
    row_mask = dist < np.mean(dist) + multiple * np.std(dist)

    filtered_matches = matches[row_mask]
    
    filtered_moving_points =  moving_points[row_mask,:]
    filtered_reference_points = reference_points[filtered_matches, :]
    
    return [filtered_moving_points, filtered_reference_points]
```

### Compute rotation R and translation t to minimize E(R,t)

```python
def ComputeTransform(filtered_moving_points, filtered_reference_points):
    
    m1_points_mean = np.mean(filtered_reference_points, axis=0) 
    m2_points_mean = np.mean(filtered_moving_points, axis=0) 
    
    m1_difference = filtered_reference_points - m1_points_mean
    m2_difference = filtered_moving_points - m2_points_mean 
    
    #change q to 3xn and then multiply by p
    A = np.transpose(m2_difference) @ m1_difference
    
    #take SVD  and return R, assuming v is already transposed,
    u, s, vh = np.linalg.svd(A)
    v = np.transpose(vh)
    uT = np.transpose(u)
    
    detMat = np.eye(3)
    detMat[2,2] = np.linalg.det(v@uT)
    R = v @ detMat @ uT
    
    #Compute translation
    T = np.transpose(m1_points_mean) - R @ np.transpose(m2_points_mean)
    
    Hom_transform = np.eye(4, 4)
    Hom_transform[:3, :3] = R
    Hom_transform[:3, 3] = T
    
    return Hom_transform
```

###  Compute Error as mean of the distances between points in $p_i$ and $q_i$

```python
def getError(filtered_moving_points, filtered_reference_points):
    
    error = np.mean(abs(filtered_moving_points-filtered_reference_points))
    
    return error
```

### Iterate after scan alignment: qi ← Rqi + t

```python
def point_to_point_ICP(reference_mesh, moving_mesh, error_threshold=10, loop_count=20, bad_point_std = 2):
    reference_points = reference_mesh.vertices

    tree = cKDTree(reference_points)
    new_mesh = moving_mesh.copy()
    error_list = []
    
    for i in range(loop_count):
        moving_points = new_mesh.vertices
        dist, matches = match_points(tree, moving_points)

        filtered_moving_points, filtered_reference_points = reject_bad_points(dist, matches,reference_points, moving_points, boundary_ratio = 0.2, multiple = bad_point_std)
        
        Hom_transform = ComputeTransform(filtered_moving_points, filtered_reference_points, )

        new_mesh.apply_transform(Hom_transform)

        error = getError(filtered_moving_points, filtered_reference_points)
        error_list.append(error)
        print(error)

        if error < error_threshold:
            break
            
    return [new_mesh,error_list]
```

### rotate a mesh, around the z axis

```python
def get_rotated_mesh(mesh, angle):
    '''
    Input angle in degrees, sampled points
    '''
    new_mesh = mesh.copy()
    Hom_transform = np.eye(4)
    
    rad_angle = angle * np.pi / 180
    
    Rot = np.array([[math.cos(rad_angle), -math.sin(rad_angle), 0], 
                    [math.sin(rad_angle), math.cos(rad_angle), 0], 
                    [0, 0, 1]])
    
    Hom_transform[:3, :3] = Rot
    
    new_mesh = new_mesh.apply_transform(Hom_transform)
    return new_mesh
```

### add noise

```python
def addNoise(mesh, noise_level, K = 1000, noise_mean = 0):
    '''
    INPUT: mesh - mesh to be changed
           noise_level - standard deviation of gaussian noise with 
           k - arbitary constant for scaling noise to length limith
        noise_mean - mean of the noise
    OUTPUT: mesh - mesh with noise
    '''
    new_mesh = mesh.copy()
    bounding_box_dimensions = new_mesh.extents
    noise = np.random.normal(0, noise_level, new_mesh.vertices.shape)
    boundedNoise = noise * (bounding_box_dimensions/K)
    new_mesh.vertices = new_mesh.vertices + boundedNoise
    return new_mesh
```

### Perform Subsampling
Use uniform sampling without replacement to perform subsampling

```python
def subsampling(reference_points, moving_points, sampling_ratio = 10):
    '''
    INPUT : matches nx3 matrix containing [checking point index,  corresponding index, distance]
            sampling_ratio - points to be sampled
    OUTPUT: sampled_matches - sampled matches 
    '''
    num_of_points = reference_points.shape[0]
    sample_size = int(num_of_points/sampling_ratio)
    index_chosen = np.random.choice(range(num_of_points), size=sample_size ,replace=False)
    
    sampled_reference_points = reference_points[index_chosen,:]
    sampled_moving_points = moving_points[index_chosen,:]
    
    print(sampled_reference_points.shape)
    
    return sampled_reference_points, sampled_moving_points
```

```python
def subsampling_point_point_ICP(reference_mesh, moving_mesh, error_threshold=10, loop_count=20, bad_point_std = 1, sub_sampling_rate = 10):
    '''
        inputs : reference point - stationary point that the other points map to
        moving points : the point set we are moving
        error_threshold : error threshold
        loop_count: amount of times to run loop
        Output : moving_points after ICP
    '''    
    reference_points = reference_mesh.vertices
    tree = cKDTree(reference_points)
    
    new_mesh = moving_mesh.copy()
    error_list = []
    
    for i in range(loop_count):
        moving_points = new_mesh.vertices
        
        dist, matches = match_points(tree, moving_points)
        
        filtered_moving_points, filtered_reference_points = reject_bad_points(dist, matches,reference_points, moving_points, boundary_ratio = 0.2, multiple = bad_point_std)
        sampled_reference_points, sampled_moving_points = subsampling(filtered_reference_points, filtered_moving_points, sampling_ratio =sub_sampling_rate)
        Hom_transform = ComputeTransform(sampled_moving_points, sampled_reference_points)
        
        new_mesh.apply_transform(Hom_transform)

        error = getError(filtered_moving_points, filtered_reference_points)
        error_list.append(error)
        
        print(error)

        if error < error_threshold:
            break
            
    return [new_mesh, error_list]
```

### Reject bad points using normal information. 
Dot product is used to measure angle between normals and a threshold is used to remove bad points

```python
def reject_bad_points_normals(dist, matches, reference_points, moving_points, reference_normals, moving_normals, boundary_ratio = 0, multiple = 2):
    '''
        function to remove the points at the boundary, and points that are larger than 2/n times the sd + mean 
    '''
    row_mask = dist < np.mean(dist) + multiple * np.std(dist)

    filtered_matches = matches[row_mask]
    
    filtered_moving_points =  moving_points[row_mask,:]
    filtered_moving_normals = moving_normals[row_mask,:]
    
    filtered_reference_points = reference_points[filtered_matches, :]
    filtered_reference_normals = reference_normals[filtered_matches, :]
    
    
    #filter by angles.
    dot_prod = np.sum(filtered_reference_normals*filtered_moving_normals,axis=1)
    intersection_angles = np.arccos(dot_prod)
    angleMask = intersection_angles < 120 * np.pi/180
    
    filtered_moving_points =  filtered_moving_points[angleMask,:]
    filtered_moving_normals = filtered_moving_normals[angleMask,:]
    
    filtered_reference_points =  filtered_reference_points[angleMask, :]
    filtered_reference_normals = filtered_reference_normals[angleMask, :]
    
    #since the normals are all unit magnitude
    
    return [filtered_moving_points, filtered_reference_points, filtered_reference_normals]
```

### Compute approximate R and t that would minimize point to plane error
as described by "Linear Least-Squares Optimization for Point-to-Plane ICP Surface Registration" by Kok-Lim Low

```python
def getHomogeneousMatrix(x_optimal):
    r = Rotation.from_euler('xyz', x_optimal[:3], degrees= False)
    Rot = r.as_matrix()
    Trans = x_optimal[3:6]
    
    Hom_transform =  np.eye(4)
    Hom_transform[:3, :3] = Rot
    Hom_transform[:3, 3] = Trans
    
    return Hom_transform
```

```python
def ComputeTransformPointToPlane(reference_points, moving_points, normals):

    A = np.cross(moving_points, normals, axis=1)
    b = np.sum(normals*reference_points, axis = 1) - np.sum(normals *moving_points, axis=1)
    A = np.c_[A, normals]
    
    #A pseudo inverse
    A_pseudo_inverse = np.linalg.pinv(A)
    X_opt = A_pseudo_inverse @ b
    
    M = getHomogeneousMatrix(X_opt)
    
    return M
```

```python
def point_to_plane_ICP(reference_mesh, moving_mesh, error_threshold=10, loop_count=20, bad_point_std = 2):
    '''
        inputs : reference point - stationary point that the other points map to
        moving points : the point set we are moving
        error_threshold : error threshold
        loop_count: amount of times to run loop
        Output : moving_points after ICP
    '''
    reference_points = reference_mesh.vertices
    reference_normals = reference_mesh.vertex_normals
    
    error_list=[]

    tree = cKDTree(reference_points)
    new_mesh = moving_mesh.copy()
    
    for i in range(loop_count):
        moving_points = new_mesh.vertices
        moving_normals = new_mesh.vertex_normals
#         print(new_mesh.vertex_normals[0:5,:])
        dist, matches = match_points(tree, moving_points)
        
        filtered_moving_points, filtered_reference_points, filtered_reference_normals =  reject_bad_points_normals(dist, matches,reference_points, moving_points, reference_normals, moving_normals)
        
        Hom_transform = ComputeTransformPointToPlane(filtered_reference_points, filtered_moving_points, filtered_reference_normals)

        new_mesh = new_mesh.apply_transform(Hom_transform)

        error = getError(filtered_moving_points, filtered_reference_points)
        print(error)
        error_list.append(error)

        if error < error_threshold:
            break
            
    return [new_mesh, error_list]
```
